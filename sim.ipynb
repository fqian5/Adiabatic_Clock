### AA vs NOAA

```python
for n_qubits in range(3,9):  
    post_selection_list = [0,n_qubits+1]
    min_gap_noaa =[]
    min_gap_aa =[]
    g_array = np.linspace(-1,-0.0001,3)
    for g in g_array:
        new_state,unitaries_list = sequential_prep_unitary(g,n_qubits)
        U = np.array(product_unitaries_loop(unitaries_list[::-1]))
        unitaries_list = [U]
        # construct clock Ham
        clock_qubits = int(np.log2(len(unitaries_list)+1))
        H_clock = H_tot(unitaries_list).real
        H_final = np.kron(np.eye(2**clock_qubits),Hamiltonian_large_dict[n_qubits])+np.kron(plus_penalty(clock_qubits),np.eye(Hamiltonian_large_dict[n_qubits].shape[0]))
        if np.isclose(np.sum(H_final.imag.data), 0):
            H_final = H_final.real
        H_clock = scipy.sparse.dia_matrix(H_clock)
        H_final = scipy.sparse.dia_matrix(H_final)
        k = 5
        eigvals,eigvecs = scipy.sparse.linalg.eigsh(H_clock,k=k,which = 'SA')
        rescale_c = eigvals[1]-eigvals[0]
        eigvals1,eigvecs1 = scipy.sparse.linalg.eigsh(H_final,k=k,which = 'SA')
        rescale_f = eigvals1[1]-eigvals1[0]
        E0_list = []
        E1_list = []
        differece_noaa = []

        v_array = np.linspace(0,1,25)
        overlap = []
        for v in v_array:

            H_adia = v*H_final/rescale_f + (1-v)*H_clock/rescale_c

            evals,evecs=scipy.sparse.linalg.eigsh(H_adia,k=k,which = 'SA')
            diff = evals[1] - evals[0]
            differece_noaa.append(diff)
            overlap.append(np.dot(eigvecs1[:,0],evecs[:,0]))

        min_gap_noaa.append(min(differece_noaa))
        plt.plot(v_array, differece_noaa,label = 'g = '+str(g)+' noaa')  
        
        ###aa simulation
        U = aa(post_selection_list,U,1)@np.array(product_unitaries_loop(unitaries_list[::-1]))
        unitaries_list = [U]
        clock_qubits = int(np.log2(len(unitaries_list)+1))
        H_clock = H_tot(unitaries_list).real
        H_final = np.kron(np.eye(2**clock_qubits),Hamiltonian_large_dict[n_qubits])+np.kron(plus_penalty(clock_qubits),np.eye(Hamiltonian_large_dict[n_qubits].shape[0]))
        if np.isclose(np.sum(H_final.imag.data), 0):
            H_final = H_final.real

        H_clock = scipy.sparse.dia_matrix(H_clock)
        H_final = scipy.sparse.dia_matrix(H_final)
        k = 5
        eigvals,eigvecs = scipy.sparse.linalg.eigsh(H_clock,k=k,which = 'SA')
        rescale_c = eigvals[1]-eigvals[0]
        eigvals1,eigvecs1 = scipy.sparse.linalg.eigsh(H_final,k=k,which = 'SA')
        rescale_f = eigvals1[1]-eigvals1[0]
        E0_list = []
        E1_list = []
        differece_aa = []
        v_array = np.linspace(0,1,25)
        overlap = []
        for v in v_array:

            H_adia = v*H_final/rescale_f + (1-v)*H_clock/rescale_c
            evals,evecs=scipy.sparse.linalg.eigsh(H_adia,k=k,which = 'SA')
            diff = evals[1] - evals[0]
            differece_aa.append(diff)
            overlap.append(np.dot(eigvecs1[:,0],evecs[:,0]))
            gs = evecs[:,0]

        min_gap_aa.append(min(differece_aa))
        plt.plot(v_array, differece_aa,label = 'g = '+str(g)+' aa')  
        plt.xlabel('s')
        plt.ylabel('Gap') 
        plt.legend()
    data_dict['min_gap_noaa'+str(n_qubits)] = min_gap_noaa
    data_dict['min_gap_aa'+str(n_qubits)] = min_gap_aa
    plt.savefig('aa_vs_noaa'+str(n_qubits)+'_qubits.png')
    plt.show()
```
